```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pt
```

```python
col={'instant':'S_NO','dteday':'DATE','season':'SEASONS','yr':'YEAR','mnth':'MONTH','hr':'HOUR',
    'holiday':'HOLIDAY','weekday':'WEEKEND','workingday':'WORKING_DAY','weathersit':'WEATHER_SITE','temp':'TEMPRATURE','atemp':'A_TEMP','hum':'HUMIDITY',
    'windspeed':'WIND_SPEED','casual':'CASUAL','registered':'REGISTERED','cnt':'COUNT'}
```

```python
data=data.rename(columns=col)
data.head()
```

```python
data[['YEAR', 'MONTH', 'DAY']] = data['DATE'].str.split('-', expand=True)
data.head()
```

```python
new_order=['S_NO','DATE','YEAR','DAY','MONTH','SEASONS','HOLIDAY',
           'WEEKEND','WORKING_DAY','WEATHER_SITE','TEMPRATURE','A_TEMP','HUMIDITY',
            'WIND_SPEED','CASUAL','REGISTERED','COUNT']	
data=data[new_order]
data.head()
```

```python
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Autumn', 4: 'Winter'}
month_mapping = {'01': 'Jan', '02': 'Feb', '03': 'Mar', '04': 'Apr', '05': 'May', '06': 'Jun',
                 '07': 'Jul', '08': 'Aug', '09': 'Sep', '10': 'Oct', '11': 'Nov', '12': 'Dec'}

# Apply mappings
data['SEASONS'] = data['SEASONS'].map(season_mapping)
data['MONTH'] = data['MONTH'].map(month_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(X_train)

x_test = sc.transform(X_test)
```

```python
from sklearn.metrics import  r2_score
```

```python

from sklearn.linear_model import LinearRegression
model_1=LinearRegression()
model_1.fit(X_train, y_train)
ypred_li = model_1.predict(X_test)

100*r2_score(ypred_li,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_3 = RandomForestRegressor(n_estimators = 10, random_state = 0)
model_3.fit(X_train,y_train)
ypred_rf=model_3.predict(X_test)
100*r2_score(ypred_rf,y_test)
```
